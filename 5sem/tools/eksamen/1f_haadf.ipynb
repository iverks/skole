```python
import numpy as np
import matplotlib.pyplot as plt
import hyperspy.api as hs
```

```python
s = hs.load("datasett/STEM_HAADF_image.hspy")
s
```

```python
s.plot()
```

```python
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar
import matplotlib.font_manager as fm
import matplotlib.patheffects as patheffects
fontprops = fm.FontProperties(size=18)

scalebar_kwargs = {'size': 200, 'label': '200 nm', 'loc': 4, 'frameon': False, 'color': 'white', 'size_vertical': 20, 'label_top': False, 'fontproperties': fontprops}

def add_scalebar(ax: plt.Axes): 
    scalebar = AnchoredSizeBar(transform=ax.transData, **scalebar_kwargs)
    # Denne legger til et svart omriss rundt scalebar teksten, for  gjre den lettere  lese
    scalebar.txt_label._text.set_path_effects([patheffects.withStroke(linewidth=2, foreground='black', capstyle="round")])
    ax.add_artist(scalebar)
```

```python
fig, ax = plt.subplots()
cax = ax.imshow(s)
add_scalebar(ax)
ax.axis("off")
fig.savefig("1i/haadf", dpi=300, pad_inches=0, bbox_inches="tight")
```
