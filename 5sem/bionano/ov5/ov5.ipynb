```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def oppg_1():
    fig, ax = plt.subplots()
    T_MAX = 10 # s
    K_PLUS = 1 # s^-1
    K_MINUS = 1 # s^-1
    DELTA_T = 0.01 # s
    p_open = 0.8
    states = [1]
    t = 0
    while t < T_MAX:
        if np.random.random() < p_open:
            states.append(0)
        else:
            states.append(1)
        p_open += - K_PLUS * p_open + K_MINUS * (1-p_open)
        t += DELTA_T
    ax.plot(states)
    fig.show()
oppg_1()
```
